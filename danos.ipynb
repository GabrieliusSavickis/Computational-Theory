### What is Countdown?

*Countdown* originated from the French game show "Des chiffres et des lettres" (Of Numbers and Letters). The game challenges two players to compete in verbal and mathematical skills by forming the longest word possible from a random selection of letters, or achieving a target number using a set of numbers and basic arithmetic operations.[(1)](http://www.screenonline.org.uk/tv/id/1295696/index.html) 

### Rules of the Game

The game is structured with clear rules to facilitate competition:[(2)](http://datagenetics.com/blog/august32014/index.html)

1. **Tile Selection**:
   - Six face-down numbered tiles are chosen from 24 shuffled tiles.
   - Tiles are divided into two categories: Large Numbers and Small Numbers.
     - Large Numbers include {25, 50, 75, 100}.
     - Small Numbers include two of each from 1 to 10.
   - Contestants select unseen tiles from the Large set (up to 4) and the rest from the Small set to total six numbers.

2. **Gameplay**:
   - A three-digit target number is randomly generated by a computer.
   - Contestants have 30 seconds to approximate the target number using the four basic arithmetic operations.
   - Tiles can be used only once, and concatenation of numbers is not allowed.
   - No calculations may result in negative numbers or fractions.

3. **Scoring**:
   - Exact solution: 10 points.
   - Within five of the target: 7 points.
   - Within ten of the target: 5 points.

The Countdown numbers game poses a complex blend of search and optimization problems. Players must search for the right operations to yield the target number while optimizing their choice of operations based on constraints such as limited use of numbers and the requirement for results to be whole numbers. These constraints introduce significant computational challenges that define the game's inherent complexity.

### Complexity of the Numbers Game

The Countdown numbers game can be formulated as a constraint satisfaction problem (CSP), where the goal is to reach a specific target number using a combination of six chosen numbers and basic arithmetic operations. The main challenges include ensuring the integrity of operations, limiting the use of numbers, and optimizing the solution process within the given time constraints. For example, consider a scenario where the target number is 653, and the available numbers are {25, 50, 75, 100, 3, 8}. A potential solution could involve calculating \(75 \times 8 = 600\), then adding \(50 + 3 = 53\) to reach exactly 653. This example demonstrates how selecting the right operations in the right order is key to solving the problem efficiently [(6)](https://www.elsevier.com/books/programming-language-pragmatics/scott/978-0-12-374514-9).

#### Search Space Complexity
Big O notation is a mathematical notation that describes the upper limit of the time complexity or space complexity of an algorithm as the input size grows [3](https://pd.daffodilvarsity.edu.bd/course/material/book-430/pdf_content) For the Countdown numbers game, Big O notation helps in quantifying the maximum number of operations an algorithm might require to find a solution as the game's parameters change.

1. **Combining Operations and Numbers**:
   - Each game provides six numbers and four operations (addition, subtraction, multiplication, division). If we consider the game where any number can pair with any operation, and each number can only be used once per operation, the immediate potential combinations can be calculated as a factorial, specifically \(6!\) (since there are six numbers to arrange in sequence). When considering that each of these numbers can be operated with four different operations in a sequence, the complexity can further be described as \(4^6\) for the operations alone, since each number could be associated with any of the four operations.
   - Therefore, the combined complexity of choosing operations and arranging numbers can be denoted as \(O(4^6 \times 6!)\), showcasing an exponential growth in complexity as more numbers or operations are potentially added [(9)](https://books.google.ie/books?hl=en&lr=&id=MTpsAQAAQBAJ&oi=fnd&pg=PR6&dq=Sedgewick,+R.,+%26+Wayne,+K.+(2011).+Algorithms+(4th+ed.).+Addison-Wesley+Professional.&ots=QhkECJM9qT&sig=Dn75YV4zKGIVnFT4r0nr_K4XeqM&redir_esc=y#v=onepage&q&f=false).

2. **Practical Constraints Reduce Complexity**:
   - While the theoretical complexity can be very high, practical constraints such as the rules that division must result in a whole number and subtraction must not produce negative results significantly prune the search space. These constraints can reduce the feasible search space, but the worst-case scenario (all operations and numbers being valid at every step) still adheres to the calculated Big O notation.

3. **Implications of Exponential Complexity**:
   - The exponential complexity \(O(4^6 \times 6!)\) indicates that the algorithm’s time to find a solution increases dramatically with an increase in the number of operations or numbers. This complexity is critical in real-time games like Countdown where the solution must be reached within 30 seconds.
   - This exponential increase is a key reason why optimized algorithms with heuristic approaches or pruning techniques are necessary to handle the game efficiently within its time constraints.

### Solution Approach

Building on the theoretical concepts outlined earlier, this section describes the practical approach taken to solve the Countdown numbers game using Python. The solution leverages a combination of programming paradigms and computational models to develop an efficient algorithm that adheres to the game's rules and constraint

#### Algorithm Design 
The core of the solution is a recursive backtracking algorithm, which systematically explores all potential combinations of numbers and arithmetic operations to reach the target number. This approach is chosen for its flexibility and effectiveness in handling the dynamic nature of the game's constraints, such as the limitation on the use of numbers and the requirement for results to be whole numbers.

**Key Features of the Algorith**
1. **Recursive Structure:** The algorithm recursively attempts every possible operation (addition, subtraction, multiplication, division) on all available numbers. It progresses by choosing a number and an operation, applying them, and then recursively proceeding with the next number [(3)](https://pd.daffodilvarsity.edu.bd/course/material/book-430/pdf_content)

2. **Pruning:** Early termination of branches occurs when a sequence of operations cannot possibly yield the target number, enhancing efficiency by reducing unnecessary computation [(4)](https://thuvienso.hoasen.edu.vn/handle/123456789/8967)

3. **State Tracking:** We maintain the state of the current calculation and the list of numbers still available for use, ensuring that each number is used no more than once. 

**Paradigms Used**
- **Functional Programming:** Ensuring that functions are pure and side-effects are minimized, which is crucial for the clarity and predictability of the recursive calls [(14)](https://academic.oup.com/comjnl/article/32/2/98/543535).
- **Object-Orientated Programming:** Encapsulating game elements into objects and classes, such as operations and numbers, to improve modularity and manageability of the code [(15)](https://dl.acm.org/doi/abs/10.1145/1402521.1413138)



#### 1. Validation Check
**Purpose**: The valid_operation function ensures that any arithmetic operation applied during the game adheres to the specific rules of Countdown. This function checks for the validity of division and subtraction, as these operations have strict rules regarding their outcomes to ensure game integrity.

```python
from itertools import permutations

def valid_operation(a, op, b):
    if op == '/' and (b == 0 or a % b != 0):  # Check for valid division
        return False
    if op == '-' and (a - b < 0):  # Check that subtraction doesn't result in negative values
        return False
    return True
```

#### Explanation
- **Division (/):** This operation checks two conditions—whether the divisor is zero (to avoid division by zero errors) and whether the division result is a whole number (no remainders are allowed).
- **Subtraction (-):** This operation checks that the result is non-negative, adhering to the game's rules that negative results are not permissible.

#### 2. Apply Operations
**Purpose:**  The apply_operation function executes the arithmetic operation between two given numbers. It supports the four basic arithmetic operations, ensuring that each operation adheres to the standard arithmetic rules and the game's constraints.

```python
def apply_operation(a, op, b):
    if op == '+': return a + b # if the operation is a plus, it applies '+' to a and b
    if op == '-': return a - b # if the operation is subtraction, it applies '-' to a and b
    if op == '*': return a * b # if the operation is multiply, it applies '*' to and b
    if op == '/': return a // b # if the the operation is divide, it applies '/' to a and b
```

#### Explanation 
- **Addition (+):** This applies addition to the two numbers
- **Subtraction (-):** This applies subtraction to the two numbers
- **Multiplication (+):** This applies multiplication to the two numbers
- **Division (+):** This applies division to the two numbers

#### 3. Solve Numbers Function
**Purpose:** The solve_numbers function orchestrates the search for a valid sequence of operations that sum to a given target using a specific set of numbers. It leverages permutations to explore all potential orders in which the numbers can be combined, ensuring comprehensive coverage of possibilities.

```python
def solve_numbers(numbers, target):
    # Iterate over all permutations of the given numbers to try every possible sequence of numbers
    for perm in permutations(numbers):
        # Start solving recursively from the first number in the current permutation
        if solve_recursive(perm, 1, perm[0], str(perm[0])):
            return True  # Return True immediately if a solution is found
    return False  # Return False if no permutation leads to a solution
```

**Explanation**
- **Permutations:** This approach considers each permutation of the input numbers, representing every possible order in which the numbers can be arranged. This is crucial as the sequence of operations might affect the final result.This is crucial as the sequence of operations might affect the final result. There are n! (n factorial) permutations for n numbers, each representing a different sequence to explore.
- **Recursive Solving:** Initiates a recursive process that tries to combine the numbers in the current permutation using arithmetic operations to achieve the target value. Each permutation is processed through recursive calls that explore every combination of the arithmetic operations (+, -, *, /). The recursive depth and the number of operations expand exponentially with the number of numbers involved, potentially executing up to 4^n operations for each permutation due to four possible operations at each step.
- **Immediate Return:** The function returns True as soon as a valid solution is found, optimizing the search process by not continuing once the solution is identified. This early termination helps in reducing unnecessary computations, but the worst-case time complexity remains significant due to the factorial growth in permutations and exponential growth in recursive calls.

#### 4. Solve Recursive Function
**Purpose:** 'solve_recursive' is designed to determine if a sequence of operations on a permutation of numbers can result in the target value. It recursively builds an arithmetic expression from the numbers and checks if it equals the target.

```python
def solve_recursive(numbers, current_index, current_result, expression):
    # Base case: Check if all numbers have been used
    if current_index == len(numbers):
        if current_result == target:
            # If the final result matches the target, print the expression and return True
            print(f"Solution found: {expression} = {target}")
            return True
        else:
            # If the final result does not match the target, return False
            return False

    # Get the current number from the sequence
    current_number = numbers[current_index]
    
    # Attempt to add, subtract, multiply, or divide the current number with the result so far
    if solve_recursive(numbers, current_index + 1, current_result + current_number, f"({expression} + {current_number})"):
        return True
    if solve_recursive(numbers, current_index + 1, current_result - current_number, f"({expression} - {current_number})"):
        return True
    if solve_recursive(numbers, current_index + 1, current_result * current_number, f"({expression} * {current_number})"):
        return True
    if current_number != 0 and current_result % current_number == 0:  # Ensure division is valid
        if solve_recursive(numbers, current_index + 1, current_result // current_number, f"({expression} / {current_number})"):
            return True

    # If no operation results in a solution, return False
    return False

# Example usage
numbers = [25, 50, 75, 100, 3, 8]
target = 653
if not solve_numbers(numbers, target):
    print("No solution found.")  # Print a message if no solution is found
```

**Explanation**
- **Base Case:** Determines if the recursion has used all the numbers. If so, it checks if the resulting expression evaluates to the target. If it matches, it prints and returns True.
- **Recursive Operations:** Each recursion call attempts to add, subtract, multiply, or divide the current total by the next number in the sequence. This ensures all combinations of operations are tried.
- **Operation Validity:** Checks like current_number != 0 for division prevent mathematically invalid operations, aligning with game rules that disallow division by zero and require integer results.
- **Time Complexity:** O(4^n), where n is the number of numbers involved. At each step of the recursion, up to four new recursive calls are made—one for each arithmetic operation. This exponential growth in recursive calls illustrates the computational intensity of the function.
- **Space Complexity:** O(n), due to the depth of the recursive call stack which grows linearly with the number of numbers as the recursion progresses deeper for each number used.
